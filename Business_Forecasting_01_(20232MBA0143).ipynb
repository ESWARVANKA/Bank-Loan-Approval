```r
# Load necessary libraries
library(ggplot2)
library(dplyr)

# Step 1: Simulate data
set.seed(123)  # For reproducibility
market_data <- data.frame(
  Location = paste("City", 1:10),
  Demand = sample(100:1000, 10),  # Random demand values
  Competition = sample(1:20, 10),  # Random competition levels
  Infrastructure = sample(5:10, 10, replace = TRUE),  # Sampling with replacement
  ProfitMargin = round(runif(10, 5, 30), 2)  # Profit margin in percentage
)

print("Simulated Market Data:")
print(market_data)

# Step 2: Analyze the data
# Criteria: High Demand, Low Competition, High Profit Margin
filtered_data <- market_data %>%
  filter(Demand > 500, Competition < 10, ProfitMargin > 15) %>%
  arrange(desc(ProfitMargin))

print("Filtered Market Data:")
print(filtered_data)

# Step 3: Visualize the data
# Scatter plot: Demand vs. Competition
ggplot(market_data, aes(x = Competition, y = Demand, color = ProfitMargin)) +
  geom_point(size = 3) +
  labs(
    title = "Market Analysis: Demand vs Competition",
    x = "Competition Level",
    y = "Demand",
    color = "Profit Margin (%)"
  ) +
  theme_minimal()

# Highlight top markets
if (nrow(filtered_data) > 0) {
  print("Top Markets to Consider:")
  print(filtered_data$Location)
} else {
  print("No suitable markets found based on criteria.")
}
```
